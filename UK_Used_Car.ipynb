### **Metric Evaluation**

MAE (Mean Absolute Error), RMSE (Root Mean Squared Error), dan MAPE (Mean Absolute Percentage Error) adalah metrik evaluasi yang akan digunakan untuk mengukur akurasi model prediksi dalam konteks dataset Mobil Bekas Inggris. Jadi, ada tiga metrik yang akan digunakan, yaitu MAE, MAPE, dan RMSE, untuk memberikan gambaran yang komprehensif tentang kualitas prediksi model. Metrik-metrik ini akan membantu dalam mengukur sejauh mana model dapat memprediksi harga mobil bekas dengan akurat.

**MAE (Mean Absolute Error):**
MAE adalah metrik yang mengukur rata-rata kesalahan absolut antara nilai yang diprediksi oleh model dan nilai aktual dalam dataset. Semakin rendah nilai MAE, semakin kecil kesalahan rata-rata prediksi model terhadap nilai aktual. MAE dinyatakan dalam unit yang sama dengan variabel yang diprediksi, dalam hal ini, dalam Poundsterling untuk harga mobil bekas.

**RMSE (Root Mean Squared Error):**
RMSE adalah metrik yang mengukur akar kuadrat dari rata-rata kesalahan kuadrat antara nilai yang diprediksi oleh model dan nilai aktual dalam dataset. RMSE memberikan bobot lebih pada kesalahan yang lebih besar dibandingkan dengan MSE. Semakin rendah nilai RMSE, semakin kecil kesalahan rata-rata kuadrat dari prediksi model terhadap nilai aktual. RMSE juga dinyatakan dalam unit yang sama dengan variabel yang diprediksi.

**MAPE (Mean Absolute Percentage Error):**
MAPE adalah metrik yang mengukur rata-rata selisih persentase absolut antara nilai yang diprediksi oleh model dan nilai aktual dalam dataset. MAPE mengukur tingkat kesalahan prediksi dalam bentuk persentase. Semakin rendah nilai MAPE, semakin dekat prediksi model dengan nilai aktual dalam bentuk persentase, menunjukkan tingkat akurasi yang lebih tinggi. MAPE sangat berguna untuk mengevaluasi akurasi prediksi harga mobil bekas dalam konteks dataset ini.

Metrik-metrik ini membantu untuk memahami sejauh mana model dapat mendekati nilai aktual dalam perhitungan harga mobil bekas. Dengan menggunakan MAE, RMSE, dan MAPE, maka dapat mengukur kualitas prediksi model dalam berbagai cara, termasuk kesalahan absolut, akar kuadrat dari kesalahan kuadrat, dan kesalahan persentase, yang semuanya penting dalam konteks penetapan harga yang akurat untuk mobil bekas di pasar UK.

## **Modeling**

### *Encoding dan Scaling*

Agar kolom yang bersifat kategorikal dapat digunakan sebagai fitur pada model, kolom tersebut harus diubah menjadi fitur numerikal. Saya menggunakan BinaryEncoder pada kolom modeel, dan OneHotEncoder pada kolom fuelType dan transmission karena kolom tersebut memiliki 5 unique value. Yang mana masih ada pada batas jumlah maksimal dari OneHotEncoder. Selain encoding, saya juga melakukan scaling pada kolom numerikal agar nilai-nilai pada kolom tersebut besarnya jadi tidak terlalu jauh. Saya menggunakan ColumnTransformer untuk mengubah kolom tersebut, yang mana nanti dapat diaplikasikan pada pipeline model.

```python
scaler = Pipeline([
        ('scaler', RobustScaler())
    ])

transformer = ColumnTransformer([
    ('one_hot', OneHotEncoder(), ['fuelType','transmission']),
    ('binary', BinaryEncoder(), ['model','brand']),
    ('scaler', scaler, ['mileage', 'mpg', 'tax']),
], remainder='passthrough')
```

### *Split Fitur dan Target*

```python
x = df.drop(columns=['price','car_period','car_period_value']) #x sebagai fitur
y = df['price'] #y sebagai target
```

### *Split Data Train dan Test*

Pada bagian ini, saya membagi data menjadi 2 bagian, yaitu data train dan data test. Perbandingan antara data train dan data test adalah 80:20

```python
# Split
xtrain, xtest, ytrain, ytest = train_test_split(
    x, 
    y,
    test_size=0.2,
    random_state=100
)
```

### *Model Benchmarking*

Algoritma yang akan saya gunakan pada model ini adalah LinearRegression, DecisionTreeRegressor, KNeighborsRegressor, XGBRegressor, dan RandomForestRegressor. Pada proses ini, saya akan mengambil 2 model terbaik berdasarkan MAPE pada hasil Cross Validate untuk dilakukan predict ke data test dan kemudian dilanjutkan dengan melakukan hyperparameter tuning pada 2 model terpilih.

```python
linreg = LinearRegression()
dt = DecisionTreeRegressor(random_state=100)
knn = KNeighborsRegressor()
xgb = XGBRegressor()    
rfr = RandomForestRegressor(random_state=100)

model = [linreg,dt,knn,xgb,rfr]
```

```python
list_model = []
list_mae = []
list_mape = []
list_rmse = []
mape_std = []

for i in model:
    pipe_model = Pipeline([
        ('preprocessing', transformer),
        ('modeling', i) 
    ])

    # cross validation
    cv_score = cross_validate( 
        estimator= pipe_model,  # model
        X = xtrain,            # X_train
        y = ytrain,            # y_train
        cv = 5,                 # berapa fold
        scoring= ['neg_mean_absolute_error','neg_mean_absolute_percentage_error',
                  'neg_mean_squared_error','neg_root_mean_squared_error']
    )
    
    list_model.append(i)
    list_mae.append(cv_score['test_neg_mean_absolute_error'].mean()) 
    list_mape.append(cv_score['test_neg_mean_absolute_percentage_error'].mean())
    list_rmse.append(cv_score['test_neg_root_mean_squared_error'].mean())
    mape_std.append(cv_score['test_neg_mean_absolute_percentage_error'].std())
```

```python
Hasil = pd.DataFrame()
Hasil['Model'] = list_model
Hasil['MAE'] = np.negative(list_mae)
Hasil['MAPE'] = np.negative(list_mape)
Hasil['RMSE'] = np.negative(list_rmse)
Hasil['MAPE std'] = mape_std

Hasil.sort_values(by='MAPE')
```

Berdasarkan hasil cross validate, RandomForestRegressor dan KNeighborsRegressor adalah 2 model terbaik karena memiliki nilai MAPE yang terendah dibandingkan dengan model lainnya. Maka, 2 model tersebut akan digunakan untuk predict ke test set.

###  *Prediksi Data Test dengan 2 Benchmark Model Terbaik*

```python
rfr_1 = RandomForestRegressor(random_state=100)
knr_1 = KNeighborsRegressor()

predict_cv = [rfr_1,knr_1]

predict_MAE = []
predict_MAPE = []
predict_RMSE = []

for j in predict_cv:
    pipe_model = Pipeline([
        ('preprocessing',transformer),
        ('modeling',j)
    ])

    #fit
    pipe_model.fit(xtrain,ytrain)

    #predict
    ypredict = pipe_model.predict(xtest)
    predict_MAE.append(mean_absolute_error(ytest,ypredict))
    predict_MAPE.append(mean_absolute_percentage_error(ytest,ypredict))
    predict_RMSE.append(np.sqrt(mean_squared_error(ytest, ypredict)))

Hasil_predict = pd.DataFrame()
Hasil_predict['Model'] = ['RandomForestRegressor','KNeighborsRegressor']
Hasil_predict['MAE'] = predict_MAE
Hasil_predict['MAPE'] = predict_MAPE
Hasil_predict['RMSE'] = predict_RMSE
Hasil_predict.sort_values(by='MAPE')
```

Berdasarkan hasil prediksi pada test set dengan menggunakan 2 model benchmark, RandomForestRegressor memiliki performa yang lebih baik dibandingkan dengan KNeighborsRegressor karena memiliki nilai MAPE yang lebih rendah.

####  *Hyperparameter Tuning*

#####  RandomForestRegressor

Parameter pada RandomForestRegressor yang akan dituning adalah sebagai berikut : 
- Scaler
- Max Depth
- Min Samples Split
- Min Sample Leaf

```python
minmax = MinMaxScaler()
standard = StandardScaler()
robust = RobustScaler()

hyperparam_space = {
    'preprocessing__scaler__scaler': [minmax, standard, robust],
    'algorithm__max_depth' : np.arange(1,20),
    'algorithm__min_samples_split' : np.arange(2,10),
    'algorithm__min_samples_leaf' : np.arange(1,10)
}

pipe_model = Pipeline([
    ('preprocessing', transformer),
    ('algorithm', RandomForestRegressor(random_state=100))
])

randomsearch = RandomizedSearchCV(
    estimator=pipe_model,             
    param_distributions=hyperparam_space,    
    cv=5,                           
    scoring='neg_mean_absolute_percentage_error',                   # metrics
    verbose=1,
    n_iter=100,
    n_jobs=-1                      
)
```

```python
randomsearch.fit(xtrain, ytrain)
```

```python
print(randomsearch.best_score_)
print(randomsearch.best_params_)
```

#####  KNeighborsRegressor

Parameter pada KneighborsRegressor yang akan dituning adalah sebagai berikut : 
- Scaler
- N Neighbors
- P

```python
minmax = MinMaxScaler()
standard = StandardScaler()
robust = RobustScaler()

hyperparam_space = {
    'preprocessing__scaler__scaler': [minmax, standard, robust],
    'algorithm__n_neighbors' : np.arange(2,6),
    'algorithm__p' : np.arange(1,3)
}

pipe_model = Pipeline([
    ('preprocessing', transformer),
    ('algorithm', KNeighborsRegressor())
])

gridsearch = GridSearchCV(
    estimator=pipe_model,
    param_grid=hyperparam_space,    
    cv=5,                            
    scoring='neg_mean_absolute_percentage_error',                   # metrics
    verbose=1,
    n_jobs=-1                       
)
```
